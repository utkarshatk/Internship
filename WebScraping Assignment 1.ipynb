#### 2)	Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
 
 #import libraries
```

```python
# Download imdb top 250 movie's data
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
movies = soup.select('td.titleColumn')
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value')
        for b in soup.select('td.posterColumn span[name=ir]')]
 
```

```python
# create a empty list for storing movie information
list = []
```

```python
# Iterating over movies to extract each movie details
for index in range(0, len(movies)):
     
    # Separating movie into: 'place',
    # 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    data = {"place": place,
            "movie_title": movie_title,
            "rating": ratings[index],
            "year": year,
            "star_cast": crew[index],
            }
    list.append(data)
 
 ##printing movie details with its rating.
for movie in list:
    print(movie['place'], '-', movie['movie_title'], '('+movie['year'] +
        ') -', 'Starring:', movie['star_cast'], movie['rating'])
 
 
##.......##
df = pd.DataFrame(list)
df.to_csv('imdb_top_250_movies.csv',index=False)
```

```python
soup
```

```python
#Headline

headline = []
for i in soup.find_all('a',class_="LatestNews-headline"):
    headline.append(i.text)
    
headline
```

```python
#Time 

time = []
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
    
time
```

```python
# Data Frame

df = pd.DataFrame({"HEADLINE":headline, "TIME":time,"URL":url})
df
```

### 9) Write a python program to scrape mentioned details from dineout.co.in :
#### i) Restaurant name
#### ii) Cuisine
#### iii) Location
#### iv) Ratings
##### v) Image URL

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
# i) Restaurant name
titles = []
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
    
titles
```

```python
# ii) Cuisine

cuisine = []
for i in soup.find_all('div',class_="detail-info"):
    cuisine.append(i.text.split('|')[1])
    
cuisine
```

```python
# iii) Location

Location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
    
Location
```

```python
# iv) Ratings

rating = []
for i in soup.find_all('div',class_="rating-txt hide"):
    rating.append(i.text)
    
rating
```

```python
# v) Image URL

url = []
for i in soup.find_all('img',class_="no-img"):
    url.append(i['data-src'])
    
url
```

```python
df = pd.DataFrame({"Restaurant name":titles, "Cuisine":cuisine, "Location":Location, "Ratings":rating, "URL":url})
df
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
# rank
rank = []
for i in soup.find_all('td', class_="gsc_mvt_p"):
    rank.append(i.text)
    
rank
```

```python
#publication 
pub = []
for i in soup.find_all('td',class_="gsc_mvt_t"):
    pub.append(i.text)
    
pub
```

```python
#h5 Index

h5_index = []
for i in soup.find_all('a',class_="gs_ibl gsc_mp_anchor"):
    h5_index.append(i.text)
    
h5_index
```

```python
#H5 Median

h5_median = []
for i in soup.find_all('span',class_="gs_ibl gsc_mp_anchor"):
    h5_median.append(i.text)
    
h5_median
```

```python
print(len(rank),len(pub),len(h5_index),len(h5_median))
```
